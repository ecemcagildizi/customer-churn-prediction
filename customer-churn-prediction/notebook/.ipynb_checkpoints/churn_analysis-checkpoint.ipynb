```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

df = pd.read_csv('/Users/ecemcagildizi/Desktop/customer-churn-prediction/data/Churn.csv')

print(df.head())
```

```python
df.drop("customerID", axis = 1, inplace=True)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

df.isnull().sum()
```

```python
df.dropna(inplace=True)

df.isnull().sum()
```

```python
df["Churn"] = df["Churn"].map({"Yes" : 1, "No" : 0})
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())
```

```python
X = df.drop("Churn", axis=1)
y = df["Churn"].map({"Yes": 1, "No": 0})
```

```python
cat_cols = X.select_dtypes(include=["object"]).columns
num_cols = X.select_dtypes(exclude=["object"]).columns
```

```python
print(cat_cols)
print(num_cols)
```

```python
X_encoded = pd.get_dummies(
    X,
    columns=cat_cols,
    drop_first=True
)
```

```python
import sklearn
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_encoded[num_cols] = scaler.fit_transform(X_encoded[num_cols])
```

```python
X_encoded[num_cols].head()
```

```python
y = df["Churn"].replace({"Yes": 1, "No": 0})
```

```python
y.isna().sum()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
y_proba = model.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

fpr, tpr, thresholds = roc_curve(y_test, y_proba)
auc_score = roc_auc_score(y_test, y_proba)

print("AUC Score:", auc_score)
```

```python
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f"AUC = {auc_score:.3f}")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Logistic Regression")
plt.legend()
plt.show()
```

```python
feature_importance = pd.DataFrame({
    "Feature": X_encoded.columns,
    "Coefficient": model.coef_[0]
})
```

```python
feature_importance["Abs_Coefficient"] = feature_importance["Coefficient"].abs()

feature_importance_sorted = feature_importance.sort_values(
    by="Abs_Coefficient",
    ascending=False
)

feature_importance_sorted.head(10)
```

```python
top_features = feature_importance_sorted.head(10)

plt.figure(figsize=(8, 5))
sns.barplot(
    x="Coefficient",
    y="Feature",
    data=top_features
)
plt.title("Top 10 Features Influencing Customer Churn")
plt.show()
```
